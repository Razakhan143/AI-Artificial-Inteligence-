```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.columns
```

```python
data=titanic[['embarked', 'class', 'who','alive']]
```

```python
data.dropna(axis=0,inplace=True)
```

```python
X=data.iloc[:,:-1]
y=data.iloc[:,-1:]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=0, shuffle=True)
```

```python
X.columns
```

```python
X['embarked'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder
oe=OrdinalEncoder(categories=[['S', 'C', 'Q',],['Third', 'First', 'Second'],['man', 'woman', 'child']])
la=LabelEncoder()
```

```python
X_test
```

```python
X_train=oe.fit_transform(X_train)
X_test=oe.fit_transform(X_test)
y_train=la.fit_transform(y_train)
y_test=la.fit_transform(y_test)
```

```python
X_test
```

```python
data.tail(5)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
dt.score(X_test,y_test)
```

```python
tips.columns
```

```python
data=tips[['sex', 'smoker', 'day', 'time','total_bill']]
```
