```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
# ['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'fare',
#        'embarked', 'class', 'who', 'adult_male', 'deck', 'embark_town',
#        'alive', 'alone']
```

```python
data=titanic[['sex', 'age','fare','embarked', 'class', 'who','alive']]
```

```python
data.head(3)
```

```python
data.dropna(axis=0,inplace=True)
```

```python
X=data.iloc[:,:-1]
y=data.iloc[:,-1:]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=0, shuffle=True)
```

```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder,LabelEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
X_train.head(2)
```

```python
transform=ColumnTransformer(transformers=[
    ('oht',OneHotEncoder(sparse=False,dtype=np.int16),[0,3]),
    ('oe',OrdinalEncoder(categories=[['Third', 'First', 'Second'],['man', 'woman', 'child']],dtype=np.int16),[4,5]),
    # ('le',LabelEncoder(),['alive']),
    ('ss',StandardScaler(),[1,2])
    ],remainder='passthrough')
```
