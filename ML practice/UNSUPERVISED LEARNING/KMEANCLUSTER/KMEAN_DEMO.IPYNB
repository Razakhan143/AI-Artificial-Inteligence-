```python
from sklearn.cluster import KMeans
```

```python
X=df[['1.4','0.2']]
```

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(X)
    wcss.append(km.inertia_)
    
    
plt.plot(range(1,11),wcss)
```

```python
km = KMeans(n_clusters=3)
y_means = km.fit_predict(X)
```
